## Plotting Categorical Data

#### Jitter Plot

```python
df2=pd.read_csv('Automobile.csv')
```

```python
df2.head()
```

```python
sns.catplot(x='fuel_system',y='normalized_losses',data=df2)
```

```python
sns.catplot(x='fuel_system',y='normalized_losses',jitter=False,data=df2)
```

#### Hue Plot

```python
sns.catplot(x='fuel_system',y='normalized_losses',hue='number_of_doors',data=df2)
```

```python
sns.catplot(x='fuel_system',y='normalized_losses',hue='number_of_doors',kind='swarm',data=df2)
```

```python
sns.catplot(x='engine_location',y='normalized_losses',hue='number_of_doors',kind='swarm',data=df2)
```

#### Boxplot

```python
sns.catplot(x='fuel_system',y='normalized_losses',kind='box',data=df2)
```

```python
sns.catplot(x='fuel_system',y='normalized_losses',hue='number_of_doors',kind='box',data=df2)
```

#### Violin Plot

```python
sns.catplot(x='fuel_system',y='normalized_losses',hue='number_of_doors',kind='violin',data=df2)
```

```python
sns.catplot(x='fuel_system',y='normalized_losses',hue='number_of_doors',kind='violin',split=True,data=df2)
```

#### Bar Plot

```python
sns.catplot(x='fuel_system',y='normalized_losses',hue='number_of_doors',kind='bar',data=df2)
```

#### Point Plot

```python
sns.catplot(x='fuel_system',y='normalized_losses',hue='number_of_doors',kind='point',data=df2)
```

#### Swarm Plot

```python
sns.catplot(x='fuel_system',y='normalized_losses',col='fuel_type',aspect=0.9,hue='number_of_doors',kind='swarm',data=df2)
```

## Visualizing the Distribution of Dataset

### Univariate Distribution

#### Histogram

```python
sns.distplot(df2.engine_size)
```

```python
sns.distplot(df2.engine_size,kde=False,rug=True)
```

### Bivariate Distribution

```python
sns.jointplot(x='engine_size',y='price',data=df2)
```

#### Hexplot

```python
sns.jointplot(x='engine_size',y='price',kind='hex',data=df2)
```

```python
sns.jointplot(x=df2.engine_size,y=df2.price,kind='hex')
```

#### KDE Plot

```python
sns.jointplot(x='engine_size',y='price',kind='kde',data=df2)
```

#### Heatmaps

```python
corrmat=df2.corr()
f, ax = plt.subplots(figsize=(12,9))
sns.heatmap(corrmat, vmax=.8, square=True) # It is only taking  numeric columns from the dataset
```

#### Boxen Plot

```python
sns.catplot(x='engine_size',y='price',kind='boxen',height=8,aspect=2.7,hue='number_of_doors',data=df2)
```
